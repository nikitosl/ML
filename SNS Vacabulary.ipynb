```python
z = np.random.normal(size=500)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

## Численные величины

### Точки на плоскости

https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

#### Группировка

```python
sns.relplot(x='total_bill', y='tip', hue='day', data=tips)
```

#### Группировка и разделение по столбцам

```python
sns.relplot(x='total_bill', y='tip', hue='day', col='time', data=tips)
```

#### Разделение графиков на строки, колонки и ещё и группировка

```python
sns.relplot(x='total_bill', y='tip', hue='day', col='time', row='sex', data=tips)
```

#### Объединение графиков по col_wrap штук в строке

```python
sns.relplot(x='total_bill', y='tip', hue='time', col='day', col_wrap=2, data=tips)
```

#### Добавление разной формы точкам

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='time', data=tips)
```

#### Добавление разного размера точкам

```python
sns.relplot(x='total_bill', y='tip', hue='time', size='size', col='time', data=tips,
           palette=['b', 'r'], sizes=(10, 200))
```

#### Изменение размера графика

```python
sns.relplot(x='timepoint', y='signal', hue='event', col='region', data=fmri,
           height=5, aspect=1.7)
```

### График со средним и отклонением

#### Среднее значение

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci=None)
```

#### Среднее значение с 95-процентным доверительным интервалом 

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')
```

#### Среднее значение со средним квадратичным отклонением 

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci='sd')
```

## Категориальные величины

### Точки на плоскости

#### Базовый график с категориальным признаком

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

#### Без распределения вширь

```python
sns.catplot(x='day', y='total_bill', jitter=False, data=tips)
```

#### С красивым распределением вширь

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

#### С группировкой по признаку

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips);
```

#### C определённым порядком следования категориального признака

```python
sns.catplot(x='smoker', y='tip', order=['No', 'Yes', 'MayBe'], data=tips)
```

### Boxplots

#### Базовый boxplot

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips)
```

#### Boxplot  с группировкой

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='box', data=tips);
```

#### Чуть более красивый boxplot для больших датасетов

```python
sns.catplot(x='color', y='price', kind='boxen', data=diamonds);
```

### Violinplots - объединяет boxplot и оценку плотности распределения

#### Базовый violin plot

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips);
```

#### Violin plot, разбитый пополам относительно группировки

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split=True, data=tips);
```

#### Совмещаем violin plot и swarm plot

```python
g = sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips, ax=g.ax);
```

## Вероятностное распределение датасета

### distplot - Построение одномерных распределений

#### Распределение вместе с плотностью распределения

```python
sns.distplot(z);
```

#### Без плотности распределения

```python
sns.distplot(z, kde=False);
```

#### С отображением внизу значений

```python
sns.distplot(z, rug=True);
```

#### Без распределения

```python
sns.distplot(z, hist=False)
```

#### Добавляем на график идеальное распределение из пакета scipy.stats

```python
sns.distplot(z, fit=stats.t);
```

#### Базовый kde plot с заполнением

```python
sns.kdeplot(z, shade=True)
```

### jointplot - Построение двумерных распределений

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

#### Базовый jointplot

```python
sns.jointplot(x='x', y='y', data=df);
```

#### Добавляем прямую регресии

```python
sns.jointplot(x='x', y='y', data=df, kind='reg');
```

#### Joint plot с более понятной отрисовкой плотности точек

```python
sns.jointplot(x='x', y='y', kind='hex', data=df);
```

#### Вместо распределения плотность

```python
sns.jointplot(x='x', y='y', kind='kde', data=df);
```

#### Тоже самое, только цвета сглажены и очень красиво 

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.jointplot(x='x', y='y', kind='kde', data=df, cmap=cmap, n_levels=60, shade=True);
```

### Визуализация парных отношений

#### Базовый pair plot

```python
sns.pairplot(iris);
```

#### Парные взаимоотношения, только заданы подробнее и с плотностью распределения

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) # Диогональ
g.map_offdiag(sns.kdeplot, shade=True, n_levels=10); # Переферия
```

#### Визуализация некоторых парных величин и регрессионной прямой

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],
             height=5, aspect=.8, kind="reg");
```

## Визуализация регрессии

### lmplot - Для отрисовки линейной регрессии

#### Линия регрессии и 95-процентный доверительный интервал

```python
sns.lmplot(x="total_bill", y="tip", data=tips);
```

#### Для категориальный величин находим центральную тенденцию и по ней строем регрессию

Вместо

```python
sns.lmplot(x='size', y='tip', data=tips);
```

Получаем

```python
sns.lmplot(x='size', y='tip', data=tips, x_estimator=np.median);
```

### Полиномиальная регрессия

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, order=2, scatter_kws={"s": 80});
```

### Борьба с выбросами

```python
sns.lmplot(x='x', y='y', data=anscombe.query("dataset == 'III'"))
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80});
```

## Разные стили графиков

Seaborn поддерживает 3 темы для графиков:  
"darkgrid", "whitegrid",  "dark", "white", "ticks"  
Настройка темы производится командой sns.set_style(<тема>)

```python
for theme in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_style(theme)
    sns.relplot(x='total_bill', y='tip', hue='day', data=tips)
```
